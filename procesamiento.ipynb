```python
import pandas as pd
import os
```

```python
file =r'campeche/'
archivos = [f for f in os.listdir(file) if f.endswith('.xls')]
campeche = []

for a in archivos:
    path = os.path.join(file, a)
    df = pd.read_excel(path, skiprows=5)
    campeche.append(df)
campeche_df = pd.concat(campeche, ignore_index=True, join = 'inner')
campeche_df['Estado'] = 'Campeche'
campeche_df
```

```python
file =r'Quintana Roo/'
archivos = [f for f in os.listdir(file) if f.endswith('.xls')]
QRoo = []

for a in archivos:
    path = os.path.join(file, a)
    df = pd.read_excel(path, skiprows=5)
    QRoo.append(df)
    
QRoo_df = pd.concat(QRoo, ignore_index=True, join= 'inner')
QRoo_df['Estado'] = 'Quintana Roo'
QRoo_df
```

```python
file =r'Yucatán/'
archivos = [f for f in os.listdir(file) if f.endswith('.xls')]
Yuc = []

for a in archivos:
    path = os.path.join(file, a)
    df = pd.read_excel(path, skiprows=5)
    Yuc.append(df)
    
Yuc_df = pd.concat(Yuc, ignore_index=True, join= 'inner')
Yuc_df['Estado'] = 'Yucatán'
Yuc_df
```

```python
peninsula = pd.concat([campeche_df, QRoo_df, Yuc_df])
peninsula.info()
```

```python
peninsula.columns = peninsula.columns.str.lstrip()
# sort ascendent by 'Ejercicio'
peninsula = peninsula.sort_values(by='Ejercicio')
peninsula.drop(columns = ['Nombre Del Candidato(a) Independiente', 
                          'Primer Apellido Del Candidato(a) Independiente', 
                          'Segundo Apellido Del Candidato(a) Independiente'], inplace=True)
peninsula.columns
```
